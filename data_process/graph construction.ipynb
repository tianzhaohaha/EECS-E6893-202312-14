```python
path = "./data.json"
with open(path, 'r') as file:
    data = json.load(file)
```

```python
for item in data.keys():
    tmp = ''
    for title in data[item]:
        tmp=tmp+title+'; '
    data[item] = tmp
```

```python
data['AAPL']
```

```python
prompt = """
Based on the following text, find the companies appeared in it, and then give the stock symbol about these companies.
The answer should look like this: "The companies contained in the text are: AAPL, MMM."
TEXT: [ANSWER]
"""
```

```python
proc_data = {}
```

```python
i=0
for item in data.keys():
    i+=1
    print(i/503)
    tmp_q = Prompt.replace('[ANSWER]', data[item])

    response = openai.Completion.create(
      engine="text-davinci-003",
      prompt=tmp_q,
      max_tokens=50
    )

    proc_data[item] = response.choices[0].text.strip().replace("The companies contained in the text are: ",'').split(', ')
```

```python
# 写入 JSON 到文件
with open('./proc_data.json', 'w') as f:
    json.dump(proc_data, f)
```
